## Primera funci√≥n

```python
%autoreload 2

from src import ODE_euler  
import numpy as np
import matplotlib.pyplot as plt

def f(t, y):
    return y - t**2 + 1

def f_real(t):
    return (t+1)**2 - 0.5*np.exp(t)

N = 10  
a, b = 0, 2 
y_t0 = 0.5  

resultado = ODE_euler(a=a, b=b, f=f, y_t0=y_t0, N=N)
resultado5 = ODE_euler(a=a, b=b, f=f, y_t0=y_t0, N=5)
resultado20 = ODE_euler(a=a, b=b, f=f, y_t0=y_t0, N=20)

t_euler = resultado[1]
y_euler = resultado[0]

t_euler20 = resultado20[1]
y_euler20 = resultado20[0]

t_euler5 = resultado5[1]
y_euler5 = resultado5[0]

t_exacta = np.linspace(a, b, 400)
y_exacta = f_real(t_exacta)

plt.plot(t_exacta, y_exacta, label="Soluci√≥n Exacta", color='b', linestyle='--')
plt.plot(t_euler, y_euler, label="Soluci√≥n Aproximada N = 10 (Euler)", marker='o', color='r')
plt.plot(t_euler5, y_euler5, label="Soluci√≥n Aproximada N = 5 (Euler)", marker='o', color='y')
plt.plot(t_euler20, y_euler20, label="Soluci√≥n Aproximada N = 20 (Euler)", marker='o', color='purple')


plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Soluci√≥n Exacta vs Aproximada (M√©todo de Euler)')
plt.legend()
plt.grid(True)
plt.show()
```

## Segunda funci√≥n

```python
def f(t, y):
    return y/t - (y/t)**2

def f_real(t):
    return t/(1+np.log(t))

N = 10  
a, b = 1, 2 
y_t0 = 1 

resultado = ODE_euler(a=a, b=b, f=f, y_t0=y_t0, N=N)
resultado5 = ODE_euler(a=a, b=b, f=f, y_t0=y_t0, N=5)
resultado20 = ODE_euler(a=a, b=b, f=f, y_t0=y_t0, N=20)

t_euler = resultado[1]
y_euler = resultado[0]

t_euler20 = resultado20[1]
y_euler20 = resultado20[0]

t_euler5 = resultado5[1]
y_euler5 = resultado5[0]

t_exacta = np.linspace(a, b, 400)
y_exacta = f_real(t_exacta)

plt.plot(t_exacta, y_exacta, label="Soluci√≥n Exacta", color='b', linestyle='--')
plt.plot(t_euler, y_euler, label="Soluci√≥n Aproximada N = 10 (Euler)", marker='o', color='r')
plt.plot(t_euler5, y_euler5, label="Soluci√≥n Aproximada N = 5 (Euler)", marker='o', color='y')
plt.plot(t_euler20, y_euler20, label="Soluci√≥n Aproximada N = 20 (Euler)", marker='o', color='purple')


plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Soluci√≥n Exacta vs Aproximada (M√©todo de Euler)')
plt.legend()
plt.grid(True)
plt.show()
```
