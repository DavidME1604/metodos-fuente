## Diverge Jacobi - Converge Seidel (1,1)

```python
%autoreload 2
from src import gauss_jacobi, gauss_seidel

A = [[2.0, 10], [3, 2]]

b = [16, 11.0]

x0 = [1, 1]

solucion, tray = gauss_jacobi(A=A, b=b, x0=x0, tol=1e-10, max_iter=50)

solucion1, tray1 = gauss_seidel(A=A, b=b, x0=x0, tol=1e-10, max_iter=50)

```

```python
print('La solucion con Jacobi es', solucion)
print('La solucion con Seidel es', solucion1)
```

```python
import matplotlib.pyplot as plt

x_vals = [punto[0] for punto in tray]
y_vals = [punto[1] for punto in tray]

x_vals1 = [punto[0] for punto in tray1]
y_vals1 = [punto[1] for punto in tray1] 

# Crear la grfica
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, '-o', label="Trayectoria", markersize=5)
plt.plot(x_vals1, y_vals1, '-o', label="Trayectoria", markersize=5)
plt.scatter(solucion[0], solucion[1], color='red', label="Solucin Final", zorder=5)
plt.scatter(solucion1[0], solucion1[1], color='pink', label="Solucin Final", zorder=5)
plt.scatter(x0[0], x0[1], color='yellow', label="Aproximacion inicial", zorder=5)
plt.xlim(-40, 5)  # Ajusta segn tus datos
plt.ylim(0, 25)
plt.title("Trayectoria del Mtodo de Gauss-Jacobi")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()
plt.show()
```
